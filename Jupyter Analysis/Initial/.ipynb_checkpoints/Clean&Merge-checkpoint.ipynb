# Merge all csvs

### Data processing

#### Join Files

```python
merge_countries = pd.concat([order_delhi,
    order_johannesburg,
order_seoul,
order_madrid,
order_merced,
order_nauen,
order_lalbagh,
order_la,
order_sbrisbane,
order_vancouver,
order_wuhan])

merge_countries

merge_countries.fillna(0)
```

```python
# Check date format
merge_countries.info()
```

```python
# Assign datetime to date column
merge_countries['date'] = merge_countries['date'].astype('datetime64[ns]')

merge_countries.info()
```

#### Sort

```python
merge_countries = merge_countries.sort_values(by=['location', 'date'], ascending=True)

merge_countries.reset_index(drop=True)

# Replace the "empties" in the dataset with a NaN
merge_countries.replace(" ", np.nan, inplace=True)

# Hide rows containing NaN for corresponding columns
merge_countries = merge_countries[merge_countries[' pm25'].notna()]
merge_countries = merge_countries[merge_countries[' pm10'].notna()]
merge_countries = merge_countries[merge_countries[' o3'].notna()]
merge_countries = merge_countries[merge_countries[' no2'].notna()]
merge_countries = merge_countries[merge_countries[' so2'].notna()]
merge_countries = merge_countries[merge_countries[' co'].notna()]

merge_countries
```

# Filter by date

```python
# Jan, Feb, Mar 2019
start_date = '2019-01-01'
end_date = '2019-03-30'

date_filter1 = mask = (merge_countries['date'] > start_date) & (merge_countries['date'] <= end_date)

filter1 = merge_countries.loc[date_filter1]

filter1
```

```python
# Jan, Feb, Mar 2020
start_date = '2020-01-01'
end_date = '2020-03-30'

date_filter2 = mask = (merge_countries['date'] > start_date) & (merge_countries['date'] <= end_date)

filter2 = merge_countries.loc[date_filter2]

filter2
```

#### Merge Filters 1 and 2

```python
merge_filters = pd.concat([filter1,
filter2])

merge_filters
```

#### Sort

```python
merge_filters = merge_filters.sort_values(by=['location', 'date'], ascending=True)

merge_filters.reset_index(drop=True)
```

# Save file as...

#### Merge ALL

```python
merge_countries.to_csv(filename, index=False)
```

#### Save filtered data

```python
merge_filters.to_csv(merged_filename, index=False)
```
