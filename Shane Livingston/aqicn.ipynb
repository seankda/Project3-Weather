```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import scipy.stats as st
from scipy.stats import linregress
from datetime import datetime
from pprint import pprint

city = ('Amsterdam','Bangkok','Beijing','Belgrade','Budapest','Busan','Delhi','Haarlem','Hong Kong',  
        'Kyoto', 'London', 'Mumbai', 'Osaka', 'Santiago', 'Seoul','Shanghai','Taipei','Tokyo')
token = ('6a9ae5f6bddb0e60357a9117c6ead1a524f072d8')
print(city)
```

```python
# Define base url
url_city="https://api.waqi.info/feed/"
#print(url_city)
```
