```python
# Data manipulation
import pandas as pd
import numpy as np
```

```python
# Import CSV
q1_2019_csv = "waqi-covid19-airqualitydata-2019Q1_modified.csv"
q1_2019_pd = pd.read_csv(q1_2019_csv, low_memory=False)
q1_2019_pd.head()
```

```python
q2_2019_csv = "waqi-covid19-airqualitydata-2019Q2_modified.csv"
q2_2019_pd = pd.read_csv(q2_2019_csv, low_memory=False)
q2_2019_pd.head()
```

```python
q1_2020_csv = "waqi-covid19-airqualitydata-2020_modified.csv"
q1_2020_pd = pd.read_csv(q1_2020_csv, low_memory=False)
q1_2020_pd.head()
```

```python
q1_2020_pd.count()
```

```python
grpd_2020 = q1_2020_pd.groupby(['Specie'])
grpd_2020.count()
```

```python
Q1_2020_pm25 = q1_2020_pd[q1_2020_pd["Specie"].str.contains('pm25')]
Q1_2020_pm25
```

```python
Q1_2020_pm10 = q1_2020_pd[q1_2020_pd["Specie"].str.contains('pm10')]
Q1_2020_pm10
```

```python
Q1_2020_o3 = q1_2020_pd[q1_2020_pd["Specie"].str.contains('o3')]
Q1_2020_o3
```

```python
Q1_2020_no2 = q1_2020_pd[q1_2020_pd["Specie"].str.contains('no2')]
Q1_2020_no2
```

```python
Q1_2020_co = q1_2020_pd[q1_2020_pd["Specie"].str.contains('co')]
Q1_2020_co
```

```python
Q1_2020_so2 = q1_2020_pd[q1_2020_pd["Specie"].str.contains('so2')]
Q1_2020_so2
```

```python
q1_2020_polutants = {}
q1_2020_polutants = Q1_2020_pm25.append(Q1_2020_pm10)
q1_2020_polutants = q1_2020_polutants.append(Q1_2020_o3)
q1_2020_polutants = q1_2020_polutants.append(Q1_2020_no2)
q1_2020_polutants = q1_2020_polutants.append(Q1_2020_co)
q1_2020_polutants = q1_2020_polutants.append(Q1_2020_so2)
q1_2020_polutants
```

```python
filtered_cities_2020 = q1_2020_polutants.loc[q1_2020_polutants['City'].isin(['Amsterdam', 'Bangkok', 'Beijing', 'Belgrade', 'Bilboa', 'Budapest', 'Busan', 'Delhi', 'Haarlem', 'Hong Kong', 'Kyoto', 'London', 'Mumbai', 'Osaka', 'Santiago', 'Seoul', 'Shanghai', 'Taipei', 'Tokyo', 'WrocÅ‚aw'])]
filtered_cities_2020
```

```python
pivot_data_2020 = pd.pivot_table(filtered_cities_2020, columns=['Specie'], index=['Date', 'City', 'Country'])
pivot_data_2020 = pivot_data_2020.dropna()
pivot_data_2020
```

```python
pivot_data_2020.to_csv("pivot_2020.csv")
```

```python
new_q1_2020_csv = "pivot_2020_modified.csv"
new_q1_2020_pd = pd.read_csv(new_q1_2020_csv, low_memory=False)
new_q1_2020_pd.head()
```

```python
grpd_2020_city = new_q1_2020_pd.groupby(['City'])
comp2020 = grpd_2020_city.count().sort_values('Date', ascending=False).head(150)
comp2020.to_csv('comp2020.csv')
comp2020
```
